#### Importing Python Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option("display.max_columns", None)
sns.set_style('whitegrid')
RANDOM_STATE = 42
```

#### Loading the Dataset

```python
df = pd.read_csv('../Dataset/Customer_Churn.csv')
df.head()
```

#### Finding out the Shape, Number of Rows and Columns in the Dataset

```python
print(f'Shape of the Dataset : {df.shape}')
print(f'Number of Rows in the Dataset : {df.shape[0]}')
print(f'Number of Columns in the Dataset : {df.shape[1]}')
```

#### Displaying the Complete Information about the Dataset

```python
df.info()
```

#### Displaying the Complete Statistical Information about the Dataset

```python
df.describe(include='all')
```

#### Stripping the Whitespaces in Feature Names

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

#### Correcting the Datatypes of the Dataset

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors='coerce')
```

#### Finding and Correcting the Missing Values in the Dataset

```python
print(f'Number of Missing Values : \n{df.isna().sum()}')
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
print(f'Number of Missing Values : \n{df.isna().sum()}')
```

#### Finding out the number of Duplicate values|

```python
print(df.duplicated().sum())
```

#### Finding out the Outliers in the Dataset

```python
def outliers(col):
    q1, q3 = np.percentile(col.dropna(), [25, 75])  # handle NaNs if present
    iqr = q3 - q1
    lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    return col[(col < lower) | (col > upper)]

for column in ['tenure', 'MonthlyCharges', 'TotalCharges']:
    print(column, 'outliers:')
    print(outliers(df[column]))
    print()
```

#### Encoding the Binary Yes/No Columns

```python
yes_no_cols = [
    "Partner", 
    "Dependents", 
    "PhoneService", 
    "PaperlessBilling",
    "SeniorCitizen"
]
yes_no_map = {
    "Yes" : 1,
    "No" : 0
}
```

```python
df[yes_no_cols] = df[yes_no_cols].replace(yes_no_map)
df['gender'] = df.gender.map({'Male' : 1, 'Female' : 0})
df.Churn = df.Churn.map(yes_no_map)
df.head()
```

#### Plotting the Target Column Distribution

```python
sns.countplot(x=df.Churn)
plt.title('Overall Churn Rate')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

#### Plotting the Numerical Columns in the Dataset

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

#### Plotting the Categorical Columns in the Dataset

```python
cat_cols = ["Contract", "InternetService", "PaymentMethod", "MultipleLines"]
for col in cat_cols:
    plt.figure(figsize=(6,3))
    sns.countplot(y=col, data=df,hue=col, order=df[col].value_counts().index)
    plt.title(col)
    plt.tight_layout()
    plt.show()
```

#### Plotting the Numerical Columns vs Target Variable

```python
for col in num_cols:
    plt.figure()
    sns.boxplot(x="Churn", y=col, data=df)
    plt.title(f"{col} vsChurn")
    plt.show()
```

#### Plotting the Categorical Columns vs Target Column

```python
for col in cat_cols:
    ct = pd.crosstab(df[col], df["Churn"], normalize="index")
    ct.plot(kind="bar", stacked=True, figsize=(6,3))
    plt.title(f"Churn rate by {col}")
    plt.ylabel("Proportion")
    plt.show()
```

#### Plotting the Numerical Columns Relationships

```python
sns.scatterplot(
    x=df['tenure'],
    y=df['MonthlyCharges'],
    hue=df['Churn'].map({0: 'No', 1: 'Yes'}),
    alpha=0.7
)
plt.title('Charges vs Tenure by Churn')
plt.tight_layout()
plt.show()
```

#### Plotting the Correlation Heat Map

```python
corr = df[num_cols].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Numeric Feature Correlations")
plt.tight_layout()
plt.show()
```

#### Churn Rate Across Customer Tenure (in Months)

```python
tenure_curve = df.groupby("tenure")["Churn"].mean()
tenure_curve.plot(figsize=(8,3))
plt.title("Churn Rate by Tenure (months)")
plt.ylabel("Churn Rate")
plt.xlabel("Tenure")
plt.show()
```
